## List of y values

In the next cell, we plot the function $ y = e^x $ where $e \approx 2.718$.

$$ y = \sum_{i=0}^{\infty} \frac{x^i}{i!} $$

```python
x = np.linspace(0.0, 3.0, 1001)

y = np.exp(x)

fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x, y)
```
